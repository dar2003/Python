**The .notnull() method in Pandas is used to detect non-missing (non-null) values in a DataFrame or Series. It returns a boolean object (of the same shape) where each element is True if the value is not null (NaN or None), and False if the value is missing.**

```python
example['price'].notnull()
```

**The .isnull() method in Pandas is used to detect missing (null) values in a DataFrame or Series. 
It identifies entries with missing values, such as NaN or None, and returns a boolean object of the same shape**

```python
example[example['price'].isnull()]
```
