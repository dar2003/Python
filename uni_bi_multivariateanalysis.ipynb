```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
table = pd.DataFrame(iris.data,columns=iris.feature_names)
table.head()
```

```python
table['target'] = iris.target
table.head()
```

```python
table.shape
```

```python
setosa=table.loc[table['target']==0]
```

```python
setosa
```

```python
virginica=table.loc[table['target']==1]
virginica
```

```python
versicolor=table.loc[table['target']==2]
versicolor
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.plot(setosa['sepal width (cm)'],np.zeros_like(setosa['sepal width (cm)']),'s')
plt.plot(virginica['sepal width (cm)'],np.zeros_like(virginica['sepal width (cm)']),'o')
plt.plot(versicolor['sepal width (cm)'],np.zeros_like(versicolor['sepal width (cm)']),'d')
plt.xlabel('sepal width')
```

**OBSERVATION FROM ABOVE GRAPH 
Setosa: The Setosa species has the narrowest sepal width, with the data points clustered around the 2.5 cm mark.
Versicolor: The Versicolor species shows a wider range of sepal widths, with the data points distributed between approximately 2.5 cm and 3.5 cm.
Virginica: The Virginica species exhibits the widest sepal widths, with the data points ranging from around 3.0 cm to 4.5 cm.**

```python
plt.plot(setosa['sepal length (cm)'],np.zeros_like(setosa['sepal length (cm)']),'s')
plt.plot(virginica['sepal length (cm)'],np.zeros_like(virginica['sepal length (cm)']),'o')
plt.plot(versicolor['sepal length (cm)'],np.zeros_like(versicolor['sepal length (cm)']),'d')
plt.xlabel('sepal length')
```

**OBSERVATIONS FROM ABOVE GRAPH:
setosa: The Setosa species has the shortest sepal length, with the data points clustered around the 5.0 cm mark.
Versicolor: The Versicolor species shows a wider range of sepal lengths, with the data points distributed between approximately 5.5 cm and 7.0 cm.
Virginica: The Virginica species exhibits the longest sepal lengths, with the data points ranging from around 6.0 cm to 8.0 cm.**

```python
plt.plot(setosa['petal length (cm)'],np.zeros_like(setosa['petal length (cm)']),'s')
plt.plot(virginica['petal length (cm)'],np.zeros_like(virginica['petal length (cm)']),'o')
plt.plot(versicolor['petal length (cm)'],np.zeros_like(versicolor['petal length (cm)']),'d')
plt.xlabel('petal length')
```

```python
plt.plot(setosa['petal width (cm)'],np.zeros_like(setosa['petal width (cm)']),'s')
plt.plot(virginica['petal width (cm)'],np.zeros_like(virginica['petal width (cm)']),'o')
plt.plot(versicolor['petal width (cm)'],np.zeros_like(versicolor['petal width (cm)']),'d')
plt.xlabel('petal width')
```
